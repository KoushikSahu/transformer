```python
import torch
import torch.nn as nn
import torchinfo
```

```python
transformer_model = nn.Transformer()
```

```python
transformer_model
```

```python
src = torch.rand((10, 32, 512))
tgt = torch.rand((20, 32, 512))
out = transformer_model(src, tgt)
out.shape
```
